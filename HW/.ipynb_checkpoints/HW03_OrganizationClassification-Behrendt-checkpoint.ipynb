## Problem 2: Binary Classification with Logistic Regression (20 points)

Now let's move to classification with four points in 2D:
- Class 0: (1,0), (2,0)
- Class 1: (3,1), (4,1)

Think about the encoding: what values will your $y_i$ take? 

Let's write $x_1$ for the x-coordinate and $x_2$ for the $y$-coordinate. Our model computes the probability:
$$P(y=1|\mathbf{x}) = \sigma(w_0 + w_1x_1 + w_2x_2)$$

where $\sigma(z)$ is the sigmoid function:
$$\sigma(z) = \frac{1}{1 + e^{-z}}.$$

We will use the loss function (binary cross-entropy):
$$L(w_0,w_1,w_2) = -\frac{1}{4}\sum_{i=1}^4 [y_i\log(p_i) + (1-y_i)\log(1-p_i)]$$
where $p_i = \sigma(w_0 + w_1x_{1i} + w_2x_{2i})$.

## Problem 3: Finding and Visualizing the Decision Boundary (20 points)

#### The Model

In logistic regression for 2D classification, we model the probability that $y=1$ given input coordinates $(x_1,x_2)$:
$$P(y=1|\mathbf{x}) = \sigma(w_0 + w_1x_1 + w_2x_2)$$
where œÉ is the sigmoid (logistic) function:
$$\sigma(z) = \frac{1}{1 + e^{-z}}.$$

#### The Loss Function
For binary classification with labels $y ‚àà \{0,1\}$, we use binary cross-entropy loss:
$$
L(w_0,w_1,w_2) = -\frac{1}{n}\sum_{i=1}^n [y_i\log(p_i) + (1-y_i)\log(1-p_i)]
$$
where $p_i = P(y=1|x_i) = œÉ(w_0 + w_1x_{1i} + w_2x_{2i})$

#### The Decision Boundary
- The decision boundary occurs where $P(y=1|x) = 0.5$
- Since $\sigma(0) = 0.5$, this happens when $w_0 + w_1x_1 + w_2x_2 = 0$
- This creates a line in the $x_1-x_2$ plane with:
  - slope = $-w_1/w_2$
  - intercept = $-w_0/w_2$

üóíÔ∏è **Tasks: 3.1 (1 points)** Given four points:
- Class 0: (1,0), (2,0)
- Class 1: (3,1), (4,1)

Find the "best" parameters for this model using **grid search** written in Python (no `sklearn`, `numpy` or other libraries!!). A helper code is given below if you wish to use it. (In reality we might use something more efficient, like a gradient descent algorithm, but I want this to be very robust, clear and focussed.)

````python
### ANSWER

# helper code if you want to use it
# if you use this, heavily comment it so that you can do it yourself next time

import numpy as np
import matplotlib.pyplot as plt

# Data
X = np.array([[1,0], [2,0], [3,1], [4,1]])  # Each row is [x‚ÇÅ,x‚ÇÇ]
y = np.array([0, 0, 1, 1])


def compute_loss(w0, w1, w2, X, y):
    """Compute binary cross-entropy loss."""
    z = w0 + w1*X[:,0] + w2*X[:,1]  # X[:,0] is x‚ÇÅ, X[:,1] is x‚ÇÇ
    p = 1/(1 + np.exp(-z))  # sigmoid
    eps = 1e-15  # avoid log(0)
    return -np.mean(y * np.log(p + eps) + (1-y) * np.log(1 - p + eps))

# Grid search over parameters
#   grid search is easy to understand, easy to code and ensures you cover the span of the parameters
#   but, it can be slow and inefficient
w0_range = np.linspace(-10, 10, 20)
w1_range = np.linspace(-10, 10, 20)
w2_range = np.linspace(-10, 10, 20)
best_loss = float('inf')
best_w0, best_w1, best_w2 = 0, 0, 0

# Brute-force grid search add your code below
def grid_search():
    """
    Performs a grid search to find the best w0, w1, and w2.
    """
    best_loss = float('inf')
    best_params = {'w0': None, 'w1': None, 'w2': None}


    print("Starting grid search...")
    # Iterate through every possible combination of parameters
    for w0_guess in w0_range:
        for w1_guess in w1_range:
            for w2_guess in w2_range:
                # Calculate the loss for the current combination
                loss = compute_loss(w0_guess, w1_guess, w2_guess, X,y)

                # If this is the best loss we've seen, save the parameters
                if loss < best_loss:
                    best_loss = loss
                    best_params['w0'] = w0_guess
                    best_params['w1'] = w1_guess
                    best_params['w2'] = w2_guess
    
    print("Grid search finished.")
    return best_params, best_loss


best_parameters, minimum_loss = grid_search()




print(f"\nBest Parameters Found:")
print(f"w0 = {best_parameters['w0']}")
print(f"w1 = {best_parameters['w1']}")
print(f"w2 = {best_parameters['w2']}")
print(f"Resulting in a minimum loss of: {minimum_loss:.4f}")
````

The one-vs-rest approach trains three separate binary classifiers:
- Adelie vs rest: $P(y=0|x)$ vs $P(y\neq 0|x)$
- Gentoo vs rest: $P(y=1|x)$ vs $P(y\neq 1|x)$

$$L_k(w) = -\frac{1}{n}\sum_{i=1}^n [y_{ki}\log(p_{ki}) + (1-y_{ki})\log(1-p_{ki})]$$

````python
### ANSWER
def plot_decision_regions(X, y, clf, title):
    # Create meshgrid
    x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
    y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                        np.arange(y_min, y_max, 0.02))
    
    # Get predictions
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    # Plot
    plt.figure(figsize=(10, 8))
    plt.contourf(xx, yy, Z, alpha=0.4)
    plt.scatter(X[:, 0], X[:, 1], c=y, alpha=0.8)
    plt.title(title)
    plt.xlabel('Standardized Bill Length')
    plt.ylabel('Standardized Bill Depth')
    return plt


plot = plot_decision_regions(X_scaled, y, clf, title='SGDClassifier Decision Regions for Penguin Species')
plot.show()

````

````python
### ANSWER
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Load and prepare data
penguins = sns.load_dataset("penguins")
penguins.dropna(subset=['bill_length_mm', 'bill_depth_mm', 'species'], inplace=True)
X = penguins[['bill_length_mm', 'bill_depth_mm']]
y_labels = penguins['species']
le = LabelEncoder()
y = le.fit_transform(y_labels) # 0:Adelie, 1:Gentoo, 2:Chinstrap
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# 1. Adelie vs. Rest
y_adelie = (y == 0).astype(int)
clf_adelie = SGDClassifier(loss='log_loss', max_iter=1000, random_state=42)
clf_adelie.fit(X_scaled, y_adelie)

# 2. Gentoo vs. Rest
y_gentoo = (y == 1).astype(int)
clf_gentoo = SGDClassifier(loss='log_loss', max_iter=1000, random_state=42)
clf_gentoo.fit(X_scaled, y_gentoo)

# 3. Chinstrap vs. Rest
y_chinstrap = (y == 2).astype(int)
clf_chinstrap = SGDClassifier(loss='log_loss', max_iter=1000, random_state=42)
clf_chinstrap.fit(X_scaled, y_chinstrap)

# Create a meshgrid
x_min, x_max = X_scaled[:, 0].min() - 1, X_scaled[:, 0].max() + 1
y_min, y_max = X_scaled[:, 1].min() - 1, X_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

# Get predictions from all three classifiers
grid_points = np.c_[xx.ravel(), yy.ravel()]
Z_adelie = clf_adelie.predict(grid_points)
Z_gentoo = clf_gentoo.predict(grid_points)
Z_chinstrap = clf_chinstrap.predict(grid_points)

# Find disagreement regions (where more than one classifier predicts class 1)
Z_sum = Z_adelie + Z_gentoo + Z_chinstrap
Z_disagreement = (Z_sum > 1).reshape(xx.shape)

# Plot the results
plt.figure(figsize=(12, 9))

# Plot the disagreement region in yellow
plt.contourf(xx, yy, Z_disagreement, colors=['yellow'], alpha=0.5, levels=[0.5, 2.5])

# Overlay the decision boundaries as lines
plt.contour(xx, yy, Z_adelie.reshape(xx.shape), colors='blue', levels=[0.5], linestyles='--')
plt.contour(xx, yy, Z_gentoo.reshape(xx.shape), colors='red', levels=[0.5], linestyles='--')
plt.contour(xx, yy, Z_chinstrap.reshape(xx.shape), colors='green', levels=[0.5], linestyles='--')

# Plot the original data points
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=y_labels,
                palette='bright', s=100, alpha=0.8)

plt.title('Disagreement Between Adelie, Gentoo, and Chinstrap Classifiers')
plt.xlabel('Standardized Bill Length')
plt.ylabel('Standardized Bill Depth')
plt.show()





# Regions with overlapping positive classification are marked in yellow. Conflicts are resolved
# by comparing the confidence score of each positive confirmation, and taking the highest.
````

````python
import numpy as np
import pandas as pd
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder


penguins = sns.load_dataset("penguins")
penguins.dropna(subset=['bill_length_mm', 'bill_depth_mm', 'species'], inplace=True)
X = penguins[['bill_length_mm', 'bill_depth_mm']]
y_labels = penguins['species']
le = LabelEncoder()
y = le.fit_transform(y_labels)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train the full multiclass classifier
clf = SGDClassifier(loss='log_loss', max_iter=1000, random_state=42)
clf.fit(X_scaled, y)

# --- Predict the probability for the new point ---
# Create the point and scale it using the same scaler
new_point = np.array([[0, -1]])
new_point_scaled = scaler.transform(new_point)

# Use predict_proba to get the probabilities
probabilities = clf.predict_proba(new_point_scaled)

# Get the class with the highest probability
predicted_class_index = np.argmax(probabilities)
predicted_class_name = le.inverse_transform([predicted_class_index])[0]

# --- Print the results ---
print(f"Point to predict (original): {new_point[0]}")
print(f"Point to predict (scaled): {new_point_scaled[0].round(2)}\n")

print("Class Probabilities:")
for i, class_name in enumerate(le.classes_):
    print(f"  - {class_name}: {probabilities[0][i]:.4f} ({(probabilities[0][i] * 100):.2f}%)")

print(f"\nFinal Prediction: **{predicted_class_name}**")
````
