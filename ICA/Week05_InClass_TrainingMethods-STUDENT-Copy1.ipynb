Also, this code can be used for some of the plotting tasks you will have.

````python
def plot_learning_curves(history):
    """Plot training and validation learning curves."""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
    
    # Plot accuracy
    ax1.plot(history.history['accuracy'], label='Training')
    ax1.plot(history.history['val_accuracy'], label='Validation')
    ax1.set_title('Model Accuracy')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Accuracy')
    ax1.legend()
    
    # Plot loss
    ax2.plot(history.history['loss'], label='Training')
    ax2.plot(history.history['val_loss'], label='Validation')
    ax2.set_title('Model Loss')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Loss')
    ax2.legend()
    
    plt.tight_layout()
    plt.show()

def plot_decision_boundary(model, X, y, title):
    """Plot decision boundary for 2D data."""
    # Use PCA to reduce to 2D for visualization
    from sklearn.decomposition import PCA
    pca = PCA(n_components=2)
    X_2d = pca.fit_transform(X)
    
    # Create a mesh grid
    x_min, x_max = X_2d[:, 0].min() - 1, X_2d[:, 0].max() + 1
    y_min, y_max = X_2d[:, 1].min() - 1, X_2d[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                        np.arange(y_min, y_max, 0.1))
    
    # Make predictions for each point in the mesh
    Z = model.predict(pca.inverse_transform(
        np.c_[xx.ravel(), yy.ravel()]
    ))
    Z = np.argmax(Z, axis=1)
    Z = Z.reshape(xx.shape)
    
    # Plot the decision boundary
    plt.figure(figsize=(10, 8))
    plt.contourf(xx, yy, Z, alpha=0.4)
    scatter = plt.scatter(X_2d[:, 0], X_2d[:, 1], c=y, alpha=0.8)
    plt.colorbar(scatter)
    plt.title(title)
    plt.show()
````

### Quick Review of L1 Regularization
L1 regularization adds the following term to the loss function:

$$ \lambda \sum_{i=1}^n |w_i| $$

where:
- $\lambda$ (lambda) is the regularization strength
- $w_i$ represents individual model weights
- $|\cdot|$ represents absolute value

The full loss function becomes:

$$ L_{total} = L_{original} + \lambda \sum_{i=1}^n |w_i| $$

#### Key Properties
- Promotes sparsity (many weights become exactly 0)
- Acts as a feature selector
- Can help prevent overfitting
- Different from L2 which shrinks weights toward 0 but rarely makes them exactly 0

### Quick Review of L2 Regularization
L2 regularization adds the following term to the loss function:

$$ \lambda \sum_{i=1}^n w_i^2 $$

where:
- $\lambda$ (lambda) is the regularization strength
- $w_i$ represents individual model weights
- The squared term ensures all penalties are positive

The full loss function becomes:

$$ L_{total} = L_{original} + \lambda \sum_{i=1}^n w_i^2 $$

#### Key Properties
- Penalizes large weights more heavily (quadratic penalty)
- Encourages weights to be small but not exactly zero
- Also called "weight decay" or "ridge regression"
- Helps prevent any single feature from having too much influence

### Quick Review of Early Stopping
Early stopping is a regularization technique that prevents overfitting by monitoring validation performance. Training stops when:

$$ L_{validation}(t + p) \geq L_{validation}(t) - \delta $$

where:
- $t$ is the current epoch
- $p$ is the patience parameter
- $\delta$ is the minimum improvement required (min_delta)
- $L_{validation}$ is the validation loss

#### Key Properties
- Stops training when validation performance stops improving
- Acts as a form of regularization by preventing overfitting
- Two key hyperparameters: patience and min_delta
- Saves the best model weights seen during training
